```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

```python
prev_log = 0
logs = []
count = []
for i in range (100):
    cur_log = np.log(i)
    logs.append(cur_log)
    count.append(i)
    # print(i, ' : ', cur_log, ' : ', (cur_log - prev_log))
    prev_log = cur_log


plt.figure(figsize=(16,8))
plt.xlabel('Integers')
plt.ylabel('Log(Integers)')
plt.title('Natural Logarithms 1 - 100')
plt.scatter(count, logs)
plt.show()
```

```python
# Load dataset into DataFrame variable
df = pd.read_csv('kc_house_data.csv')

# For this post, we're only looking at sqft_above as an indicator
r_skew = pd.Series(df.sqft_above)
plt.hist(r_skew)
plt.show()
print('Kurtosis: ', r_skew.kurtosis())
print('Skewness: ', r_skew.skew())
```

```python
# Log transform the right skewed data
r_skew_log = np.log(r_skew)
plt.hist(r_skew_log)
plt.show()
print('Kurtosis: ', r_skew_log.kurtosis())
print('Skewness: ', r_skew_log.skew())
```
